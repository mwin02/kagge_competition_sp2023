# Simple Gradient Descent Demo

```python
import numpy as np, matplotlib.pyplot as plt
```

For this demo, we want to minimize a function $f(x) = (x-2)^2 + 1$
- gradient = $2 * (x - 2)$

```python
def f(x):
  return (x - 2)**2 + 1

def df_dx(x):
  return 2 * (x - 2)

def do_gradient_descent(start, stepsize, iterations):
  x = start
  f_x = [start]
  for i in range(iterations):
    x -= df_dx(x) * stepsize
    f_x.append(x)
  return f_x

def plot_results(values, title=""):
  if title:
    plt.title(title)
  plt.xlim(-5, 9)
  plt.ylim(-5, 50)
  plt.xlabel("x")
  plt.ylabel("f(x)")
  plt.plot(np.linspace(-5, 9, 50), f(np.linspace(-5, 9, 50)),
           c="red", ls='-', linewidth=0.5, label="f(x)")
  plt.plot(values, [f(v) for v in values], c="blue", ls="-",
           markevery=1, marker="d", markersize=4, label="gradient descent")
  plt.plot(values[0], f(values[0]), marker="*", markersize=12, c="cyan")
  plt.legend()
  plt.show()
```

## Choosing a Step Size

```python
stepsizes = [0.005, 0.1, 0.5, 0.75, 1.1]
for stepsize in stepsizes:
  out = do_gradient_descent(5, stepsize, 10)
  plot_results(out, f"Step Size = {stepsize}")
```
